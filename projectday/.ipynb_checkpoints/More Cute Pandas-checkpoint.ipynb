The basic workflow (from http://nbviewer.ipython.org/github/cs109/content/blob/master/lec_04_wrangling.ipynb) is as follows:

1. Build a DataFrame from the data (ideally, put all data in this object)
2. Clean the DataFrame. It should have the following properties:

    * Each row describes a single object
    * Each column describes a property of that object
    * Columns are numeric whenever appropriate
    * Columns contain atomic properties that cannot be further decomposed
    
3. Explore global properties. Use histograms, scatter plots, and aggregation functions to summarize the data.
4. Explore group properties. Use groupby and small multiples to compare subsets of the data.

This process transforms your data into a format which is easier to work with, gives you a basic overview of the data's properties, and likely generates several questions for you to followup in subsequent analysis.

```python
df = pd.read_csv("cleaned.csv")
df.head()
```

```python
# find nulls and put them into an empty string
df.isbn[df.isbn.isnull()]=""
df.genre_urls[df.genre_urls.isnull()]=""
df.genres[df.genres.isnull()]=""
```

```python
# decribe tells us the properties of the dataframe
df.describe()
```

```python
# find the decades with integer division
decade =  (df.year // 10) * 10
print type(decade)
decade
```

```python
# create a bydecade dataframe and add a decade column. Didn't actually need to create a new dataframe - could have 
# added the column to the existing dataframe
bydecade_df = df[['name', 'year']]
bydecade_df['decade'] = decade

bydecade_df.head()
```

```python
# group books by decade column
dfdecade=df.groupby(decade)
print type(dfdecade)
dfdecade.groups
```

```python
decade_ratings = df.groupby(decade).rating
print type(decade_ratings)
decade_mean=decade_ratings.mean()
decade_mean.name = 'Decade Mean'
print decade_mean
```

```python
dfdecade
```

```python
decade_std=decade_ratings.std()
```

```python
decade_mean
```

```python
# the 2* does 2 times the standard deviation for a 95% confidence value. Originally we didn't have that in the 
# fill_between command
plt.plot(decade_mean.index, decade_mean.values, 'o-',
        color='r', lw=3, label='Decade Average')
plt.scatter(df.year, df.rating, alpha=.1, lw=0)
plt.fill_between(decade_mean.index, (decade_mean + 2*decade_std).values,
                 (decade_mean - 2*decade_std).values, color='r', alpha=.2)
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.legend()
plt.xlim([1900, 2015]);
```

```python
# create a set of unique genres - brings together unique words. Update function updates set. 
# A column for each unique genre with a pipe between.
genres = set()
for m in df.genres:
    genres.update(g for g in m.split('|'))
genres = sorted(genres)
print genres[-20:] # last 20
print len(genres)
```

```python
# once you split up the genres, is the genre in there or not for each row in the dataframe - a list inside a list
for genre in genres:
    df["genre:"+genre] = [genre in g.split('|') for g in df.genres]
         
df.head()
```

```python
genres_wanted=dfg.index[dfg.values> 200]
genres_wanted
```

```python
# there are 60 genres with at least 200 books in them
genres_wanted.shape, type(genres_wanted)
```

```python
fig, axes = plt.subplots(nrows=10, ncols=6, figsize=(12, 20), 
                         tight_layout=True)
#print axes.ravel()
print genres_wanted
bins = np.arange(1950, 2013, 3)
for ax, genre in zip(axes.ravel(), genres_wanted):
    #print genre, df[df[genre] == True].year.values
    ax.hist(df[df[genre] == True].year.values, 
            bins=bins, histtype='stepfilled', normed=True, color='r', alpha=.2, ec='none')
    ax.hist(df.year, bins=bins, histtype='stepfilled', ec='None', normed=True, zorder=0, color='#cccccc')
    
    ax.annotate(genre.split(':')[-1], xy=(1955, 3e-2), fontsize=12)
    ax.xaxis.set_ticks(np.arange(1950, 2013, 30))
    ax.set_yticks([])
    ax.set_xlabel('Year')
```

```python
fig, axes = plt.subplots(nrows=10, ncols=6, figsize=(12, 20), 
                         tight_layout=True)
#print axes.ravel()
bins = np.arange(2, 5, 0.1)
for ax, genre in zip(axes.ravel(), genres_wanted):
    #print genre, df[df[genre] == True].year.values
    ax.hist(df[df[genre] == True].rating.values, 
            bins=bins, histtype='stepfilled', normed=True, color='r', alpha=.1, ec='none')
    ax.hist(df.rating, bins=bins, histtype='stepfilled', ec='None', normed=True, zorder=0, color='#cccccc')
    
    ax.annotate(genre.split(':')[-1], xy=(2, 3e-1), fontsize=12)
    ax.xaxis.set_ticks(np.arange(2, 5, 1))
    ax.set_yticks([])
    ax.set_xlabel('Rating')
```
