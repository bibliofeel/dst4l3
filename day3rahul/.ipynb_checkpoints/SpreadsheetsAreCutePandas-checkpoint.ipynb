##Vectorization

The python numerical stack is based on a concept called vectorization.

Whats the idea? It is that cpu's have some ability to carry out multiple things at one time, and we should use it to increase speed. But its also something else: there is a simplicity of notation. For example, see the following

```python
# Here's a list. Go element by element and print the product.
a=[1,2,3,4,5]
b=[1,2,3,4,5]
print type(a)
for ele in a:
    print(ele*ele)
for j in range(len(a)):
    print a[j]*b[j]
```

```python
a=np.array(a)
b=np.array(b)
print a*b
```

Much simpler and much faster, as its implemented in C

### More cleaning

```python
# books where the year is null - sometimes scraping might not have worked well. Can put date into spreadsheet
# by hand or just not use these .  NaN means "not a number"
df[df.year.isnull()]
```

```python
# so using df for a new dataframe - python will clean up old unused dataframe
df = df[df.year.notnull()]
```

```python
df.shape
```

```python
df.rating_count=df.rating_count.astype(int)
df.review_count=df.review_count.astype(int)
df.year=df.year.astype(int)
```

```python
df.genre_urls[df.genre_urls.isnull()]=""
df.isbn[df.isbn.isnull()]="no isbn"
```

```python
# map takes every item of a list, applies a function to each item, and stores it as a new variable
# a 1-to-1 mapping. In this ex, for booktype getting rid of the "goodreads:" from the strings in the series; splitting 
# the string on the colon.  
# edit author urls and get rid of all but actual name.  Split on slashes. Then split on dot. then recombine in case there
# are more dots. Store in a new column called author
df['author']=df.author_url.map(lambda x: ".".join(x.split('/')[-1].split('.')[1:]))
df['booktype']=df.booktype.map(lambda x: x.split(':')[-1])
```

```python
# to see 1st row
df.author_url[0]
```

```python
df.genre_urls[0]
```

```python
# similar thing here to get the last genre element and store in a new column called genres
def saj(item):
    genres=item.split('|')
    genres=[e.split('/')[-1] for e in genres]
    return "|".join(genres)

df['genres']=df.genre_urls.map(saj)
df.head()
```

```python
decade =  (df.year // 10) * 10
print type(decade)
decade
```

```python
bydecade_df = df[['name', 'year']]
bydecade_df['decade'] = decade

bydecade_df.head()
```

```python
dfdecade=df.groupby(decade)
print type(dfdecade)
dfdecade.groups
```

```python
decade_ratings = df.groupby(decade).rating
decade_mean=decade_ratings.mean()
decade_mean.name = 'Decade Mean'
print decade_mean
```

```python
decade_std=decade_ratings.std()
```

```python
plt.plot(decade_mean.index, decade_mean.values, 'o-',
        color='r', lw=3, label='Decade Average')
plt.scatter(df.year, df.rating, alpha=.04, lw=0)
plt.fill_between(decade_mean.index, (decade_mean + decade_std).values,
                 (decade_mean - decade_std).values, color='r', alpha=.2)
plt.xlabel("Year")
plt.ylabel("Score")
plt.legend(frameon=False)
plt.xlim([1940, 2010]);
```

```python
plt.plot(decade_mean.index, decade_mean.values, 'o-', lw=2, alpha=0.7, label='Decade Average')
plt.scatter(df.year, df.rating, alpha=.04, lw=0)
plt.fill_between(decade_mean.index, (decade_mean + decade_std).values,
                 (decade_mean - decade_std).values, alpha=.1)
plt.plot(decade_mean.index, [df.rating.mean()]*len(decade_mean.values), alpha=0.6, label='Mean rating')
plt.xlabel("Year")
plt.ylabel("Score")
plt.legend(frameon=False)
plt.xlim([1940, 2010]);
```
